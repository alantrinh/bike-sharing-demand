```python
def add_date_time_features(dataframe):
    dataframe['year'] = dataframe.index.year
    dataframe['month'] = dataframe.index.month_name()
    dataframe['day'] = dataframe.index.day_name()
    dataframe['hour'] = dataframe.index.hour.astype(str) + 'h'

add_date_time_features(X_train)
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder
```

```python
X_train_time_only = X_train[['year', 'month', 'day', 'hour', 'holiday']]

one_hot_months = pd.get_dummies(X_train_time_only['month'])
one_hot_days = pd.get_dummies(X_train_time_only['day'])
one_hot_hours = pd.get_dummies(X_train_time_only['hour'])

X_ohe_features = one_hot_months.join(one_hot_days).join(one_hot_hours)
X_train_time_only = X_train_time_only.join(X_ohe_features)
X_train_time_only.drop(columns=['month', 'day', 'hour'], axis=1, inplace=True)
X_train_time_only
```

```python
m = RandomForestClassifier(max_depth=15, n_estimators=100)
m.fit(X_train_time_only, y_train)
y_pred = m.predict(X_train_time_only)
pd.Series(y_pred)
```

```python
from sklearn.metrics import mean_squared_log_error

mean_squared_log_error(y_train, y_pred, squared=False)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
categorical_steps = [('onehot', OneHotEncoder(handle_unknown='ignore'))]

categorical_transformer = Pipeline(categorical_steps)
preprocessor = ColumnTransformer(transformers=[
    ('categorical', categorical_transformer, ['month', 'day', 'hour'])
], remainder='passthrough')

final_steps = [('preprocessor', preprocessor),
     ('clf', m)] # take model from above

pipeline = Pipeline(final_steps)
pipeline.fit(X_train[['year', 'month', 'day', 'hour', 'holiday']], y_train)
```
